##### a. Distribusi Data

```python
# Histogram Distribusi data

def plot_histograms(df):
    
    # Atur size dari figure
    plt.figure(figsize=(15, 10))
    
    # List kolom yang akan diplot
    columns = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 
               'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 
               'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 
               'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'TotalSpending']
    
    # Loop untuk membuat histogram
    for i, column in enumerate(columns, 1):
        plt.subplot(5, 4, i)
        sns.histplot(df[column])
    
    # Mengatur layout agar tidak tumpang tindih
    plt.tight_layout()
    plt.show()

plot_histograms(df)
```

##### b. Analisis Distribusi Data (Uji Normalitas)

```python
# Kolmogorov Smirnov Test

from scipy.stats import kstest

def kolmogorov_smirnov_test(df):
    columns = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 
               'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 
               'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 
               'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'TotalSpending']
    
    for column in columns:
        stat, pval = kstest(df[column], 'norm')
        print(f'Kolom {column} Distribusi tidak normal (Tolak H0)') if pval < 0.05 else print(f'Kolom {column} Distribusi normal (Terima H0)')

kolmogorov_smirnov_test(df)
```

```python
# D'Agustino Pearson
from scipy.stats import normaltest

for kolom in numerik.columns:
    stat, pval = normaltest(numerik[kolom])
    if pval < 0.05:
        print(f'Distribusi {kolom} tidak normal (Tolak H0)')
    else:
        print(f'Distribusi {kolom} normal (Terima H0)')
```

<div class="alert alert-info">
INSIGHT:

* Income: Normal (Terima H0).
* Recency, MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds, NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases, NumWebVisitsMonth, Age, Total_Spending: Tidak normal (Tolak H0).

##### c. Outlier

```python
# Boxplot

def box_plots(df):
    
    # Atur size dari figure
    plt.figure(figsize=(15, 10))
    
    # List kolom yang akan diplot
    columns = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 
               'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 
               'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 
               'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'TotalSpending']
    
    # Loop untuk membuat histogram
    for i, column in enumerate(columns, 1):
        plt.subplot(5, 4, i)
        sns.boxplot(df[column])
    
    # Mengatur layout agar tidak tumpang tindih
    plt.tight_layout()
    plt.show()

box_plots(df)

```

```python
# IQR 1.5
for nama_kolom in numerik.columns:
    q3 = numerik[nama_kolom].quantile(0.75)
    q1 = numerik[nama_kolom].quantile(0.25)
    iqr = q3 - q1
    batas_atas = q3 + (iqr * 1.5)
    batas_bawah = q1 - (iqr * 1.5)
    jumlah_outliers = len(numerik[(numerik[nama_kolom] > batas_atas) | (numerik[nama_kolom] < batas_bawah)])
    print(f'Jumlah Outliers {nama_kolom}: {jumlah_outliers}')
```

```python
# Z-Score

for kolom in numerik.columns:
    std = numerik[kolom].std()
    mean = numerik[kolom].mean()

    # menghitung zscore
    zscore = abs((numerik[kolom] - mean) / std)

    # mencari outliers
    outliers = len(numerik[zscore > 3])
    print(f'Jumlah outliers {kolom} : {outliers}')
```

<div class="alert alert-info">
INSIGHT:

**Outlier Signifikan:**
   - Income: Outlier ekstrem di atas $600,000 (rata-rata $52,247).
   - MntWines, MntMeatProducts, Total_Spending: Outlier tinggi (misalnya, Total_Spending hingga ~$2,500 vs. median ~$500).

Outlier tersebut perlu dilakukan pengecekan lebih lanjut untuk menentukan apakah data tersebut valid atau tidak.

Frekuensi =

IF [Num Web Purchases] + [Num Catalog Purchases] + [Num Store Purchases] >= 10 THEN 3
ELSEIF [Num Web Purchases] + [Num Catalog Purchases] + [Num Store Purchases] >= 5 THEN 2
ELSE 1
END

```python
# menambahkan variabel Frekuensi
df['Frequency'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df
```

```python
# menambah variabel Monetary
df['Monetary'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df
```

```python
# mengambil kolom yang dibutuhkan RFM
df_rfm = df[['ID', 'Recency', 'Frequency', 'Monetary']]
df_rfm
```

```python

# Membuat Variabel RFM

def rfm_score(baris):
    # recency
    if baris['Recency'] <= 30: #Q1
        r = 3
    elif baris['Recency'] <= 60: # Q2
        r = 2
    else:
        r = 1

    # fequency
    if baris['Frequency'] < 5: #Q1
        f = 1
    elif baris['Frequency'] < 10: # Q2
        f = 2
    else:
        f = 3

    # monetary
    if baris['Monetary'] <= 500: #Q1
        m = 1
    elif baris['Monetary'] <= 1000: # Q2
        m = 2
    else:
        m = 3

    return pd.Series({'R_Score': r, 'F_Score': f, 'M_Score': m})

df_rfm[['R_Score', 'F_Score', 'M_Score']] = df_rfm.apply(rfm_score, axis=1)
df_rfm['RFM_Score'] = df_rfm['R_Score'] + df_rfm['F_Score'] + df_rfm['M_Score']
df_rfm
```

```python
# RFM Segmentation

def rfm_segment(row):
    if row['RFM_Score'] >= 7:
        return 'High-Value'
    elif row['RFM_Score'] >= 5:
        return 'Mid-Value'
    else:
        return 'Low-Value'
    
df_rfm['RFM_Segment'] = df_rfm.apply(rfm_segment, axis=1)
df_rfm['RFM_Segment'].value_counts()
```

```python
# buat Customer Secmentation =

# Manual Clustering using IF-ELSE Conditions
def segment_customer(row):
    recency = row['R_Score']
    frequency = row['F_Score']
    if recency == 3 and frequency == 3:
        return 'Champions'
    elif recency == 3 and frequency== 2:
        return 'Loyal Customers'
    elif recency == 3 and frequency == 1:
        return 'Potential Loyalist'
    elif recency == 2 and frequency == 3:
        return 'Need Attention'
    elif recency == 2 and frequency == 2:
        return 'New Customers'
    elif recency == 2 and frequency == 1:
        return 'Promising'
    elif recency == 1 and frequency == 3:
        return 'At Risk'
    elif recency == 1 and frequency == 2:
        return 'About To Sleep'
    elif recency == 1 and frequency == 1:
        return 'Hibernating'
    else:
        return 'Regular'

# Apply the segmentation function
df_rfm['RFM_Segment_Detail'] = df_rfm.apply(segment_customer, axis=1)
df_rfm['RFM_Segment_Detail'].value_counts()
```
